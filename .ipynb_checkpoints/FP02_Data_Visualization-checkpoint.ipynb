### Checking Outliers

```python
col = ['age', 'numscreens', 'fo_to_enrolled']

def outliers (df,col_name):
    Q1 = df[col_name].quantile(0.25)
    Q2 = df[col_name].quantile(0.50)
    Q3 = df[col_name].quantile(0.75)
    IQR = Q3-Q1 #Interquartile range
    low_bound  = Q1 - (1.5*IQR)
    up_bound = Q3 + (1.5*IQR)
    print(col_name)
    print(f'Lower Bound : {low_bound}'+'\n'
          f'Upper Bound : {up_bound}')
    print('-'*30)
    
for i in col:
    outliers(data,i)
```

```python
ot_age = data[(data['age']<4.5) | (data['age']>56.5)]
ot_nmsc = data[(data['numscreens']<-17.0) | (data['numscreens']>55.0)]
ot_fotoenr = data[(data['fo_to_enrolled']<-24.0) | (data['fo_to_enrolled']>40.0)]
```
